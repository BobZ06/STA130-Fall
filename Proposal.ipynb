**Analysis 1: Social Media Impact on Loneliness**

**Research Question**  
Is the average loneliness score higher among frequent social media users compared to the national average loneliness score?

**Variables**  
- **Social Media Usage**: `CONNECTION_social_media_visits_per_day`
  - **Measurement**: Ordinal scale indicating the average number of social media visits per day.
  - **Visualization**: Bar chart displaying the distribution of daily social media visits.
  - **Relevance**: Identifies levels of social media engagement, allowing segmentation of users by frequency.

- **Loneliness Score**: `LONELY_ucla_loneliness_scale_companionship`
  - **Measurement**: Ordinal scale measuring the frequency of feeling a lack of companionship.
  - **Visualization**: Histogram showing the distribution of loneliness scores for the companionship variable.
  - **Relevance**: Focuses on a specific aspect of loneliness, providing insights into how social media usage correlates with feelings of companionship.

**Preprocessing**  
1. **Data Import and Filtering**: Load `CSCS_data_anon.csv`, filter for cases where `REMOVE_case` is "No," and exclude rows with missing values in these specific columns.
2. **Segmentation**: Identify frequent social media users based on thresholds in `CONNECTION_social_media_visits_per_day`.

**Analysis Method**  
1. **Statistical Test**: Conduct a one-sample t-test (or non-parametric test, if necessary) to compare the loneliness score for companionship among frequent social media users with the national average.
2. **Bootstrapping**: Generate confidence intervals for the loneliness score to ensure robustness.

**Hypothesis and Expected Results**  
- **Null Hypothesis (H0)**: The average companionship loneliness score for frequent social media users is equal to the national average.
- **Alternative Hypothesis (H1)**: The average companionship loneliness score for frequent social media users is higher than the national average.

**Relevance**: If frequent social media users show higher loneliness scores, it suggests a potential relationship between social media usage and loneliness, helping to identify possible risk factors for social isolation.

**Analysis 2: Relationship Between Community Participation and Life Satisfaction**

**Research Question**  
Does participation in group exercises predict higher life satisfaction scores?

**Variables**  
- **Community Participation**: `CONNECTION_activities_group_exercise_p3m`
  - **Measurement**: Ordinal scale indicating frequency of participation in group exercise (e.g., yoga classes, cycling).
  - **Visualization**: Bar chart showing participation levels in group exercises.
  - **Relevance**: Highlights engagement in physical community activities, relevant to examining potential effects on life satisfaction.

- **Life Satisfaction**: `WELLNESS_subjective_happiness_scale_happy`
  - **Measurement**: Continuous scale representing self-reported happiness levels.
  - **Visualization**: Histogram showing the distribution of life satisfaction scores.
  - **Relevance**: Serves as a measure of well-being, providing a basis for assessing the impact of group exercise participation.

**Preprocessing**  
1. **Data Import and Filtering**: Load data, keeping rows where `REMOVE_case` is "No," and exclude rows with missing values in these two columns.
2. **Standardization**: Normalize the life satisfaction scale if necessary.

**Analysis Method**  
1. **Statistical Test**: Use simple linear regression to assess whether frequency of group exercise predicts life satisfaction scores.
2. **Bootstrapping**: Generate confidence intervals for the regression coefficient to reinforce result reliability.

**Hypothesis and Expected Results**  
- **Null Hypothesis (H0)**: There is no relationship between group exercise participation and life satisfaction (β = 0).
- **Alternative Hypothesis (H1)**: Higher participation in group exercise is associated with higher life satisfaction (β > 0).

**Relevance**: Positive results would suggest that community-based physical activities like group exercises could enhance life satisfaction, supporting initiatives to promote social and physical engagement for well-being.

**Analysis 3: Social Support Network and Mental Health**

**Research Question**  
How does the size of one's social support network relate to mental health scores?

**Variables**  
- **Social Support Network Size**: `CONNECTION_social_num_close_friends_grouped`
  - **Measurement**: Ordinal scale indicating the number of close friends.
  - **Visualization**: Bar chart showing the distribution of social network sizes among participants.
  - **Relevance**: Measures the extent of social support, allowing for analysis of its relationship with mental health.

- **Mental Health**: `WELLNESS_self_rated_mental_health`
  - **Measurement**: Ordinal scale representing self-rated mental health status.
  - **Visualization**: Box plot to display the distribution of mental health scores across different network sizes.
  - **Relevance**: Reflects participants’ mental health status, essential for examining its association with social network size.

**Preprocessing**  
1. **Data Import and Filtering**: Import data, filter rows based on `REMOVE_case`, and exclude rows with missing values in these columns.
2. **Standardization**: Standardize mental health scores if necessary for consistency across network sizes.

**Analysis Method**  
1. **Statistical Test**: Use ordinal logistic regression or Spearman correlation to evaluate the relationship between network size and mental health scores.
2. **Bootstrapping**: Calculate confidence intervals for each parameter to improve result reliability.

**Hypothesis and Expected Results**  
- **Null Hypothesis (H0)**: Social network size has no effect on mental health scores (β = 0).
- **Alternative Hypothesis (H1)**: A larger social network is associated with better mental health scores (β > 0).

**Relevance**: A significant positive association could indicate that larger social support networks contribute to better mental health, supporting strategies that promote social connections for psychological well-being.
